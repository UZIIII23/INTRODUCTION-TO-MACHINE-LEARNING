# Import Matplotlib
We usually import it with the alias plt (this is the standard convention):

````python
import matplotlib.pyplot as plt
````

## 1. Basic Plot – Line Plot

````python
# Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y)
plt.show()
````

## 2. Adding Titles, Labels and Grid

````python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y)

plt.title("Student Marks Over Days")
plt.xlabel("Days")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
````

## 3. Change Style: Color, Marker, Line Style

````python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y, 
         color='red', 
         marker='o', 
         linestyle='--', 
         linewidth=2, 
         markersize=8)

plt.title("Styled Line Plot")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(True)

plt.show()
````

Common marker styles:
- 'o' → circle
- 's' → square
- '^' → triangle
- 'D' → diamond
- '*' → star

Common line styles:
- '-'  solid
- '--' dashed
- ':'  dotted
- '-.' dash-dot

## 4. Multiple Lines in One Plot

````python
days = [1, 2, 3, 4, 5]

math = [85, 88, 82, 90, 87]
science = [78, 85, 80, 92, 88]
english = [90, 87, 91, 85, 89]

plt.plot(days, math, label='Math', marker='o')
plt.plot(days, science, label='Science', marker='s')
plt.plot(days, english, label='English', marker='^')

plt.title("Student Marks Comparison")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)

plt.show()
````

## 5. Bar Plot

````python
subjects = ['Math', 'Science', 'English', 'Urdu', 'Physics']
marks = [88, 85, 90, 78, 92]

plt.bar(subjects, marks, color=['#4CAF50', '#2196F3', '#FF9800', '#F44336', '#9C27B0'])

plt.title("Class Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 100)

plt.show()
````

## 6. Horizontal Bar Plot

````python
plt.barh(subjects, marks, color='skyblue')

plt.title("Horizontal Bar - Class Performance")
plt.xlabel("Marks")
plt.ylabel("Subjects")

plt.show()
````

## 7. Scatter Plot

````python
hours_studied = [2, 3, 5, 6, 7, 8, 9, 1, 4, 10]
exam_marks = [45, 55, 70, 78, 82, 85, 88, 40, 65, 92]

plt.scatter(hours_studied, exam_marks, 
            color='purple', 
            marker='o', 
            s=10,          # size
            alpha=0.9)      # transparency

plt.title("Study Hours vs Exam Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.grid(True)

plt.show()
````

## 8. Histogram

````python
marks = [45, 55, 60, 62, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100]

plt.hist(marks, bins=10, color='teal', edgecolor='black')

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
````

## 9. Pie Chart

````python
subjects = ['Math', 'Science', 'English', 'Urdu', 'Physics']
marks = [88, 85, 90, 78, 92]

plt.pie(marks, labels=subjects, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

plt.title("Subject-wise Marks Percentage")
plt.axis('equal')  # equal aspect ratio → perfect circle

plt.show()
````

## 10. Subplots (Multiple Plots in One Figure)

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1
axes[0, 0].plot([1,2,3,4], [10,20,25,30], 'b-o')
axes[0, 0].set_title("Line Plot")

# Plot 2
axes[0, 1].bar(['A','B','C'], [5,7,3], color='orange')
axes[0, 1].set_title("Bar Plot")

# Plot 3
axes[1, 0].scatter([1,2,3,4,5], [2,5,3,8,1], color='green')
axes[1, 0].set_title("Scatter Plot")

# Plot 4
axes[1, 1].hist([1,2,2,3,3,3,4,5,5], bins=5, color='purple')
axes[1, 1].set_title("Histogram")

plt.tight_layout()  # avoids overlapping
plt.show()
````
